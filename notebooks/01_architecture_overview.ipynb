## 4. Module 1: Threshold Interface

**Theoretical basis:** Baars (1988) — Global Workspace Theory

The Threshold Interface implements GWT's "competition for broadcast" — only sufficiently salient inputs gain access to the global workspace.

### Salience Gate Equation:
$$g_m = \sigma(W_s \cdot \mathbf{e}_m), \quad \tilde{\mathbf{e}}_m = g_m \odot W_p \cdot \mathbf{e}_m$$

Only tokens where $g_m > \tau$ (salience threshold) pass through.

````python
from kairosyn.model.threshold_interface import ThresholdInterface, SalienceGate

# Instantiate a small ThresholdInterface for demonstration
HIDDEN_DIM = 64
threshold = ThresholdInterface(
    text_dim=HIDDEN_DIM,
    vision_dim=HIDDEN_DIM,
    audio_dim=HIDDEN_DIM,
    hidden_dim=HIDDEN_DIM,
    gate_hidden_dim=32,
    salience_threshold=0.3,
)

# Simulate text embeddings
B, T = 2, 16  # batch=2, seq_len=16
text_embeds = torch.randn(B, T, HIDDEN_DIM)

# Forward pass
output, salience_scores = threshold(text_embeds)
print(f'Input shape  : {text_embeds.shape}')
print(f'Output shape : {output.shape}')
print(f'Text salience (mean): {salience_scores["text"].mean().item():.4f}')
````

````python
# Visualize salience gating across sequence positions
salience_gate = SalienceGate(HIDDEN_DIM, 32, threshold=0.3)
_, salience_vals = salience_gate(text_embeds[0])  # [T]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: salience scores per token position
sal = salience_vals.detach().numpy()
colors = ['#E15759' if s > 0.3 else '#AEC7E8' for s in sal]
axes[0].bar(range(T), sal, color=colors, edgecolor='white')
axes[0].axhline(0.3, color='red', linestyle='--', linewidth=1.5, label='Threshold τ=0.3')
axes[0].set_xlabel('Token Position')
axes[0].set_ylabel('Salience Score')
axes[0].set_title('Module 1: Salience Gate Scores\n(Red = above threshold → enters workspace)')
axes[0].legend()
axes[0].set_ylim(0, 1)

# Right: threshold effect on embedding norms
gated, _ = salience_gate(text_embeds[0])
orig_norms = text_embeds[0].norm(dim=-1).detach().numpy()
gated_norms = gated.norm(dim=-1).detach().numpy()
x = range(T)
axes[1].plot(x, orig_norms, 'o-', label='Original', color='#4E79A7', linewidth=2)
axes[1].plot(x, gated_norms, 's--', label='Gated', color='#E15759', linewidth=2)
axes[1].set_xlabel('Token Position')
axes[1].set_ylabel('Embedding L2 Norm')
axes[1].set_title('Embedding Norms: Before vs After Salience Gating')
axes[1].legend()

plt.tight_layout()
plt.show()
````

## 5. Module 2: Arche-Tempus Drive

**Theoretical basis:** Dennett (1991) narrative gravity; Dai et al. (2019) Transformer-XL

The Arche-Tempus Drive encodes a persistent "temporal self" via Temporal Narrative Embeddings (TNE):

$$\text{TNE}(t) = [\sin(\omega_i t + \phi_i),\, \cos(\omega_i t + \phi_i)]_{i=1}^{d/2} + \mathbf{v}_t$$

where $\mathbf{v}_t = W_v \cdot [\mathbf{h}_t - \mathbf{h}_{t-1},\, \mathbf{h}_t]$ is the **semantic velocity** — measuring how fast the meaning is changing.

````python
from kairosyn.model.arche_tempus import ArcheTemplusDrive

arche = ArcheTemplusDrive(
    hidden_dim=HIDDEN_DIM,
    narrative_embed_dim=32,
    max_temporal_horizon=128_000,
    num_temporal_heads=4,
)

# Test with a sequence
x = torch.randn(B, T, HIDDEN_DIM)
output_at, tce = arche(x)

print(f'Input shape          : {x.shape}')
print(f'Output shape         : {output_at.shape}')
print(f'Temporal Cont. Error : {tce:.4f}  (0.0 = perfect continuity)')

# Run a second time to see TCE in action (state drifts)
x2 = torch.randn(B, T, HIDDEN_DIM)
_, tce2 = arche(x2)
print(f'TCE after 2nd pass   : {tce2:.4f}  (measures narrative drift)')
````

````python
# Visualize temporal narrative embeddings across sequence positions
from kairosyn.model.arche_tempus import TemporalNarrativeEmbedding

tne = TemporalNarrativeEmbedding(HIDDEN_DIM, narrative_embed_dim=32, max_temporal_horizon=128_000)

positions = torch.arange(T).unsqueeze(0)  # [1, T]
encodings = tne._sinusoidal_temporal_encoding(positions)  # [1, T, 32]
enc_np = encodings[0].detach().numpy()  # [T, 32]

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Heatmap of temporal encoding
im = axes[0].imshow(enc_np.T, aspect='auto', cmap='RdBu_r', vmin=-1, vmax=1)
axes[0].set_xlabel('Token Position (time)')
axes[0].set_ylabel('Embedding Dimension')
axes[0].set_title('Module 2: Temporal Narrative Embedding\n(Each row = one sinusoidal frequency)')
plt.colorbar(im, ax=axes[0])

# Plot first 4 frequency components
for i in range(4):
    axes[1].plot(enc_np[:, i], label=f'dim {i}', linewidth=2)
axes[1].set_xlabel('Token Position')
axes[1].set_ylabel('TNE Value')
axes[1].set_title('Temporal Encoding: First 4 Sinusoidal Frequencies')
axes[1].legend()
axes[1].axhline(0, color='gray', linewidth=0.8)

plt.tight_layout()
plt.show()
````

## 7. Module 4: Recursion Lattice ⭐ (Core Innovation)

**Theoretical basis:** Hofstadter (1979) — Strange Loops

The Recursion Lattice is KAIROSYN's most novel contribution. It implements Hofstadter's strange loop computationally via **cross-layer self-referential attention**.

### Strange Loop Query:
$$\mathbf{Q}_L = W_Q \cdot [\mathbf{h}_L \,\|\, \mathbf{h}_{L-k} \,\|\, \mathbf{h}_{L-2k}]$$

At layer $L$, the model asks: *"Given what I thought at layers $L-k$ and $L-2k$, what should I attend to now?"*

### Strange Loop Gate:
$$\mathbf{h}'_L = \mathbf{h}_L + \alpha \cdot \sigma(W_g \cdot [\mathbf{h}_L, \mathbf{h}_{L-k}]) \odot \text{LoRA}(\mathbf{h}_{L-k})$$

````python
from kairosyn.model.recursion_lattice import RecursionLattice, StrangeLoopGate

lattice = RecursionLattice(
    hidden_dim=HIDDEN_DIM,
    num_lattice_layers=4,
    num_heads=4,
    recursion_depth=2,
    lora_rank=8,
    loop_gate_alpha=0.1,
)

x = torch.randn(B, T, HIDDEN_DIM)
output_rl, rcs = lattice(x)

print(f'Input shape            : {x.shape}')
print(f'Output shape           : {output_rl.shape}')
print(f'Recursive Conv. Score  : {rcs:.4f}  (1.0 = perfect convergence)')

# Show parameter count for the lattice
total_params = sum(p.numel() for p in lattice.parameters())
print(f'Lattice parameters     : {total_params:,}')
````

````python
# Visualize the strange loop: hidden state similarity across lattice layers
# We trace how each layer's representation relates to previous layers

def trace_recursion_lattice(lattice, x):
    """Trace hidden states through each lattice layer."""
    layer_states = []
    current = x
    cache = []
    
    for i, layer in enumerate(lattice.layers):
        past = cache[-lattice.recursion_depth:]
        with torch.no_grad():
            current, pre_norm = layer(current, past)
        cache.append(pre_norm)
        layer_states.append(current.detach())
    
    return layer_states

layer_states = trace_recursion_lattice(lattice, x)

# Compute cosine similarity matrix between all layer pairs
n_layers = len(layer_states)
sim_matrix = np.zeros((n_layers, n_layers))

for i in range(n_layers):
    for j in range(n_layers):
        h_i = layer_states[i].mean(dim=1)  # [B, D]
        h_j = layer_states[j].mean(dim=1)
        sim = F.cosine_similarity(h_i, h_j, dim=-1).mean().item()
        sim_matrix[i, j] = sim

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Layer similarity heatmap
im = axes[0].imshow(sim_matrix, cmap='YlOrRd', vmin=0, vmax=1)
axes[0].set_xlabel('Lattice Layer')
axes[0].set_ylabel('Lattice Layer')
axes[0].set_title('Module 4: Cross-Layer Hidden State Similarity\n(Strange Loop self-reference map)')
plt.colorbar(im, ax=axes[0])
axes[0].set_xticks(range(n_layers))
axes[0].set_yticks(range(n_layers))

# RCS across layers
rcs_per_layer = [sim_matrix[i, max(0, i-1)] for i in range(n_layers)]
axes[1].plot(range(n_layers), rcs_per_layer, 'o-',
             color='#E15759', linewidth=2.5, markersize=8)
axes[1].fill_between(range(n_layers), rcs_per_layer, alpha=0.2, color='#E15759')
axes[1].set_xlabel('Lattice Layer')
axes[1].set_ylabel('RCS (cosine sim to prev layer)')
axes[1].set_title('Recursive Convergence Score per Layer')
axes[1].set_ylim(0, 1)
axes[1].axhline(rcs, color='gray', linestyle='--', label=f'Mean RCS = {rcs:.3f}')
axes[1].legend()

plt.tight_layout()
plt.show()
````

## 8. Module 5: Mythogenic Engine

**Theoretical basis:** Jung (1959) — The Archetypes and the Collective Unconscious

The Mythogenic Engine maintains a library of **64 learnable Jungian archetypal prototypes** and learns to recognize which archetypes are active in the model's current internal state.

### Archetype Activation:
$$\mathbf{a} = \text{softmax}\left(\frac{\hat{\mathbf{q}} \cdot \hat{\mathbf{A}}^\top}{\tau_A}\right), \quad \mathbf{e}_\text{arch} = \mathbf{a}^\top \mathbf{A}$$

````python
from kairosyn.model.mythogenic_engine import MythogenicEngine, ArchetypeLibrary

engine = MythogenicEngine(
    hidden_dim=HIDDEN_DIM,
    num_archetypes=16,  # reduced for demo
    embed_dim=32,
    symbolic_vocab_size=64,
    archetype_temperature=0.8,
)

x = torch.randn(B, T, HIDDEN_DIM)
enhanced, symbolic_logits, aac = engine(x)

print(f'Input shape        : {x.shape}')
print(f'Enhanced shape     : {enhanced.shape}')
print(f'Symbolic logits    : {symbolic_logits.shape}')
print(f'AAC score          : {aac:.4f}  (1.0 = perfect symbolic alignment)')

# Show top archetypes
archetype_lib = engine.archetype_library
query = engine.input_norm(engine.input_proj(x.mean(dim=1)))  # [B, E]
_, scores = archetype_lib(query)  # [B, N_arch]

print('\nTop 5 Active Archetypes (sample 0):')
top5_idx = scores[0].argsort(descending=True)[:5]
for rank, idx in enumerate(top5_idx, 1):
    name = archetype_lib.archetype_names[idx]
    score = scores[0][idx].item()
    print(f'  {rank}. {name:<25} {score:.4f}')
````

````python
# Visualize archetype activation landscape
scores_np = scores[0].detach().numpy()  # [N_arch]
names = archetype_lib.archetype_names[:16]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart of archetype activations
sorted_idx = np.argsort(scores_np)[::-1]
sorted_scores = scores_np[sorted_idx]
sorted_names = [names[i] for i in sorted_idx]
colors_arch = plt.cm.viridis(np.linspace(0.2, 0.9, len(names)))

axes[0].barh(range(len(names)), sorted_scores[::-1],
             color=colors_arch, edgecolor='white')
axes[0].set_yticks(range(len(names)))
axes[0].set_yticklabels(sorted_names[::-1], fontsize=9)
axes[0].set_xlabel('Activation Score')
axes[0].set_title('Module 5: Jungian Archetype Activations\n(Current internal state signature)')
axes[0].set_xlim(0, sorted_scores.max() * 1.15)

# Radar chart of top 8 archetypes
top8_names = [names[i] for i in sorted_idx[:8]]
top8_scores = scores_np[sorted_idx[:8]]
top8_scores_norm = top8_scores / top8_scores.max()

angles = np.linspace(0, 2 * np.pi, 8, endpoint=False)
top8_scores_plot = np.concatenate([top8_scores_norm, [top8_scores_norm[0]]])
angles_plot = np.concatenate([angles, [angles[0]]])

ax_radar = axes[1]
ax_radar = plt.subplot(122, projection='polar')
ax_radar.plot(angles_plot, top8_scores_plot, 'o-', color='#76B7B2', linewidth=2)
ax_radar.fill(angles_plot, top8_scores_plot, alpha=0.25, color='#76B7B2')
ax_radar.set_xticks(angles)
ax_radar.set_xticklabels(top8_names, fontsize=8)
ax_radar.set_title('Top 8 Archetypes\n(Radar View)', pad=20)

plt.tight_layout()
plt.show()
````

## 10. Module 7: Continuity Engine

**Theoretical basis:** Dennett (1991) — Narrative Gravity

The Continuity Engine maintains a **persistent self-state** that evolves across inference steps.

### GRU-Style Self-State Update:
$$\mathbf{S}_t = \text{LayerNorm}(\mathbf{S}_{t-1} + W_u \cdot \mathbf{h}^{(L)}_t)$$

### Narrative Coherence Score:
$$\text{NCS} = \cos(\mathbf{S}_t, \mathbf{S}_{t-1})$$

````python
from kairosyn.model.continuity_engine import ContinuityEngine

engine_c = ContinuityEngine(
    hidden_dim=HIDDEN_DIM,
    state_dim=64,
    buffer_size=32,
    narrative_window=8,
)

# Simulate a multi-turn conversation
ncs_scores = []
state_norms = []

print('Simulating multi-turn conversation...')
print(f'{"Turn":>5} | {"NCS":>8} | {"State Norm":>12} | Interpretation')
print('-' * 65)

for turn in range(8):
    # Each turn has slightly different hidden states
    noise_level = 0.3 if turn < 4 else 1.5  # sudden topic shift at turn 4
    x_turn = torch.randn(1, T, HIDDEN_DIM) * noise_level
    
    output_c, ncs = engine_c(x_turn)
    state_norm = engine_c.self_state.norm().item()
    ncs_scores.append(ncs)
    state_norms.append(state_norm)
    
    interp = '🟢 Coherent' if ncs > 0.8 else ('🟡 Moderate' if ncs > 0.5 else '🔴 Drift!')
    print(f'{turn+1:>5} | {ncs:>8.4f} | {state_norm:>12.4f} | {interp}')
````

````python
# Visualize narrative coherence over conversation turns
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

turns = range(1, len(ncs_scores) + 1)

# NCS over turns
ncs_arr = np.array(ncs_scores)
axes[0].plot(turns, ncs_arr, 'o-', color='#59A14F', linewidth=2.5, markersize=8, label='NCS')
axes[0].fill_between(turns, ncs_arr, alpha=0.15, color='#59A14F')
axes[0].axhline(0.8, color='green', linestyle='--', alpha=0.6, label='High coherence (0.8)')
axes[0].axhline(0.5, color='orange', linestyle='--', alpha=0.6, label='Moderate (0.5)')
axes[0].axvline(4.5, color='red', linestyle=':', linewidth=2, label='Topic shift')
axes[0].set_xlabel('Conversation Turn')
axes[0].set_ylabel('Narrative Coherence Score (NCS)')
axes[0].set_title('Module 7: Narrative Coherence Over Time\n(Continuity Engine self-state stability)')
axes[0].legend(fontsize=8)
axes[0].set_ylim(-0.1, 1.1)
axes[0].set_xticks(list(turns))

# Self-state norm evolution
axes[1].plot(turns, state_norms, 's-', color='#4E79A7', linewidth=2.5, markersize=8)
axes[1].axvline(4.5, color='red', linestyle=':', linewidth=2, label='Topic shift')
axes[1].set_xlabel('Conversation Turn')
axes[1].set_ylabel('Self-State L2 Norm')
axes[1].set_title('Continuity Engine Self-State Growth\n(Dennett narrative gravity building)')
axes[1].legend()
axes[1].set_xticks(list(turns))

plt.tight_layout()
plt.show()
````

## 12. Epinoetic Metrics Explained

KAIROSYN uses five specialized evaluation metrics, plus a composite **Epinoetic Score**.

````python
# Simulate metric evolution across training phases
np.random.seed(42)

steps = np.linspace(0, 100, 50)

def smooth(arr, w=5):
    return np.convolve(arr, np.ones(w)/w, mode='valid')

# Simulate metric trajectories
ncs_raw = 0.61 + 0.18 * (1 - np.exp(-steps/30)) + np.random.randn(50)*0.02
tce_raw = 0.39 - 0.18 * (1 - np.exp(-steps/30)) + np.random.randn(50)*0.02
aac_raw = 0.52 + 0.14 * (1 - np.exp(-steps/40)) + np.random.randn(50)*0.015
msa_raw = 0.48 + 0.09 * (1 - np.exp(-steps/35)) + np.random.randn(50)*0.02
rcs_raw = 0.65 + 0.17 * (1 - np.exp(-steps/25)) + np.random.randn(50)*0.02

epinoetic = 0.25*ncs_raw + 0.25*rcs_raw + 0.20*aac_raw + 0.15*msa_raw + 0.15*(1-tce_raw)

fig, axes = plt.subplots(2, 3, figsize=(15, 9))
axes = axes.flatten()

metrics_data = [
    ('NCS — Narrative Coherence Score', ncs_raw, '#59A14F', '↑ higher = better', True),
    ('TCE — Temporal Continuity Error', tce_raw, '#E15759', '↓ lower = better', False),
    ('AAC — Abstraction Alignment', aac_raw, '#76B7B2', '↑ higher = better', True),
    ('MSA — Multimodal Synchrony', msa_raw, '#F28E2B', '↑ higher = better', True),
    ('RCS — Recursive Convergence', rcs_raw, '#4E79A7', '↑ higher = better', True),
    ('Epinoetic Score (composite)', epinoetic, '#B07AA1', '↑ higher = better', True),
]

# Phase boundaries
phase_boundaries = [33, 66]  # SFT | PPO | MAML

for ax, (title, data, color, direction, higher_better) in zip(axes, metrics_data):
    ax.plot(steps, data, color=color, linewidth=2, alpha=0.4)
    if len(data) >= 5:
        s_data = smooth(data)
        s_steps = steps[2:-2]
        ax.plot(s_steps, s_data, color=color, linewidth=2.5, label='Smoothed')
    for pb in phase_boundaries:
        ax.axvline(pb, color='gray', linestyle='--', alpha=0.5, linewidth=1)
    ax.text(16, ax.get_ylim()[0] + 0.01, 'SFT', color='gray', fontsize=8)
    ax.text(49, ax.get_ylim()[0] + 0.01, 'PPO', color='gray', fontsize=8)
    ax.text(82, ax.get_ylim()[0] + 0.01, 'MAML', color='gray', fontsize=8)
    ax.set_title(f'{title}\n({direction})', fontsize=10, fontweight='bold')
    ax.set_xlabel('Training Steps (%)')
    ax.set_ylim(data.min() - 0.05, data.max() + 0.05)

plt.suptitle('KAIROSYN-1 Evaluation Metrics Across Three Training Phases',
             fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('kairosyn_metrics.png', dpi=150, bbox_inches='tight')
plt.show()
print('Figure saved: kairosyn_metrics.png')
````
